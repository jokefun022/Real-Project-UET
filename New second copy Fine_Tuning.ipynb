```python
import pandas as pd

df = pd.read_csv("/content/Complete Data With Emoji.csv")
display(df.head())
display(df.info())
```

## Feature extraction

### Subtask:
Implement the feature extraction methods discussed in the research paper. This may involve techniques like TF-IDF, word embeddings, or other methods mentioned in the paper.


**Reasoning**:
Implement TF-IDF feature extraction as per the instructions.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(df['Tweet_Text_With_Emoji'])
print(tfidf_features.shape)
```

## Machine learning model implementation

### Subtask:
Implement and train the Machine Learning models discussed in the paper using the extracted features. This might include models like SVM, Naive Bayes, Logistic Regression, etc.


**Reasoning**:
Split the data into training and testing sets, import the necessary machine learning models, and initialize them.



```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(tfidf_features, df['Label'], test_size=0.2, random_state=42)

svm_model = SVC()
nb_model = GaussianNB()
lr_model = LogisticRegression()
```

**Reasoning**:
Train the initialized models using the training data.



```python
svm_model.fit(X_train, y_train)
nb_model.fit(X_train.toarray(), y_train)
lr_model.fit(X_train, y_train)
```

## Deep learning model implementation

### Subtask:
Implement and train the Deep Learning models discussed in the paper. This could involve models like LSTMs, GRUs, or CNNs.


**Reasoning**:
Implement and train a Deep Learning model (LSTM) for text classification using the preprocessed text data. This involves building the model architecture, compiling it, and training it on the training data.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SpatialDropout1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=5000, split=' ')
tokenizer.fit_on_texts(df['Tweet_Text_With_Emoji'].values)
X = tokenizer.texts_to_sequences(df['Tweet_Text_With_Emoji'].values)
X = pad_sequences(X)

embed_dim = 128
lstm_out = 192

model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=embed_dim, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.4))
model.add(LSTM(lstm_out, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(5, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

Y = pd.get_dummies(df['Label']).values
X_train_dl, X_test_dl, Y_train_dl, Y_test_dl = train_test_split(X, Y, test_size=0.2, random_state=42)

batch_size = 32
epochs = 5

model.fit(X_train_dl, Y_train_dl, epochs=epochs, batch_size=batch_size, verbose=1)
```
